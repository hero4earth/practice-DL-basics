# Linear Regression for Tensor Flow Practice

- 라이브러리 import

```python
import tensorflow as tf
```

#### 주어진 데이터 

```python
x_data = [1, 2, 3, 4]
y_data = [2, 3, 4, 5]
```

### 1. 그래프(Graph) 만들기 
#### Placeholder 지정

```python
X = tf.placeholder(tf.float32, shape=[None], name="X")
Y = tf.placeholder(tf.float32, shape=[None], name="Y")
```

#### 최적화할 변수 지정(W, b)

```python
W = tf.Variable(tf.zeros([1]), name="weight")
b = tf.Variable(tf.zeros([1]), name="bias")
```

#### H(x) =  X*W + b 

```python
hypothesis = X*W+b
```

#### 손실함수 만들기(Cost Function)

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
learning_rate = 0.1
```

#### 경사 하강법(Gradient Descent Algorithm)을 통해 최소값 찾기
- 1) 직접 작성

```python
# cost function 미분값(경사도)
gradient_W = tf.reduce_mean(((W*X+b) - Y)* X)
gradient_b = tf.reduce_mean(((W*X+b) - Y))
# 
descent_W = W - learning_rate * gradient_W
descent_b = b - learning_rate * gradient_b
update_W = W.assign(descent_W)
update_b = b.assign(descent_b)
```

- 2) 경사 하강법 최적화 함수 사용(Gradient Descent Optimizer)

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate)
train_step = optimizer.minimize(cost) # 경사 하강법의 최솟값 호출
```

### 2. 그래프(Graph) 실행
#### 그래프 실행을 위한 세션 생성
- 세션에서 모든 동작의 실행은 tf.run()으로 작동

```python
sess = tf.Session()
```

#### 변수 초기화

```python
sess.run(tf.global_variables_initializer())
```

#### 그래프 반복 실행

- Tensor Flow Gradient Descent Optimizer 함수 사용

```python
for step in range(100):
    _, cost_val = sess.run([train_step, cost], feed_dict={X: x_data, Y: y_data})
    print(step, cost_val, sess.run(W), sess.run(b))
```

```python
for step in range(100):
    _, _, cost_val = sess.run([update_W, update_b, cost], feed_dict={X: x_data, Y: y_data})
    print(step, cost_val, sess.run(W), sess.run(b))
```

#### 세션 종료
- 세션을 종료해야 그래프가 초기화가 됨

```python
sess.close()
```
